```python
from matplotlib import pyplot as plt
import math

z = [27, 31, 43, 58, 69, 86, 102, 111, 122, 137, 18, 176]
T = [71, 64, 52, 41, 33, 23, 17, 12, 2, 0, 87, -5]
```

```python
plt.scatter(z, T)
```

```python
# y = Ce^(kx) or T(deg. C) = C*e^(kZ)
C = 2.718281828**5.1554
k = -0.0275239

x = z[:]
y = [C*math.exp(k*x) for x in x]
```

```python
y
```

```python
plt.scatter(z, T, color='g')
plt.scatter(x, y, color='r')
plt.show()
```

```python
x1 = [x for x in range(0, 176)]
y1 = [C*math.exp(k*x1) for x1 in x1]

plt.scatter(z, T, color='m')
plt.plot(x1, y1)
plt.show()
```
